# 1. MiniGame on Number Guessing:
# Hint:
#     a. Choose random number between 1 to 10, 1 to 100 or any range
#     b. Give user a hint to guess the number one by one
#     c. Hint can be <,>,+,-,*,/ or anything
#     c. you also need function to compare the inputted function with guessed number
#     d. Finally, calculate the number of attempts that user taking to guess the number 
# condition: Only three attempts allowed for the user to find the generated number

```python
import numpy as np
Gen_num = np.random.randint(1,100)
inp = int(input('Enter number between 1 to 100:'))

if(Gen_num == inp):
    attempt = 0
    attempt = attempt + 1
    print('Guessed correct number',end = '\n')
    print('Own',attempt)

else:
    n = 0
    attempt = 0
    while(attempt<3):
        
        if(n==0):
            if(inp>Gen_num):
                print('Clue:1 - Guessed number is larger than generated number')
            else:
                print('Clue:1 - Guessed number is smaller than generated number')
        
            print('Guessed number is incorrect',end = '\n')
            print('please try again')
            inp = int(input('Enter the Input:'))
            if(inp==Gen_num):
                attempt = attempt + 1
                print('Guessed correct number',end = '\n')
                print(inp,end = '\n')
                print('In',attempt,'attemps you guessed correctly')
                break
            else:
                n=n+1
                attempt = attempt + 1
                
        if(n==1):
            if(inp>Gen_num):
                print('Clue:1 - Guessed number is larger than generated number')
            else:
                print('Clue:1 - Guessed number is smaller than generated number')

            if(Gen_num%2==0):
                print('Clue:2 - Number is Even')
            else:
                print('Clue:2 - Number is Odd') 
            print('Guessed number is incorrect',end = '\n')
            print('please try again')
            inp = int(input('Enter the Input:'))
            if(inp==Gen_num):
                attempt = attempt + 1
                print('Guessed correct number',end = '\n')
                print(inp,end = '\n')
                print('In',attempt,'attemps you guessed correctly')
                break
            else:
                n=n+1
                attempt = attempt + 1
        
        if(n==2):
            if(inp>Gen_num):
                print('Clue:1 - Guessed number is larger than generated number')
            else:
                print('Clue:1 - Guessed number is smaller than generated number')

            if(Gen_num%2==0):
                print('Clue:2 - Number is Even')
            else:
                print('Clue:2 - Number is Odd') 
            
            if((inp-Gen_num)!=0):
                print('Clue:3 - difference between input and Generated number is:',inp-Gen_num)
            else:
                print('Clue:3 - difference between input and Generated number is:0')
            print('Guessed number is incorrect',end = '\n')
            print('please try again')
            inp = int(input('Enter the Input:'))
            if(inp==Gen_num):
                attempt = attempt + 1
                print('Congrats,Guessed correct number',end = '\n')
                print(inp,end = '\n')
                print('In',attempt,'attemps you guessed correctly')
                break
            else:
                n=n+1
                attempt = attempt + 1
    else:
        print('Generated number is:',Gen_num)
        print('Have attended your maximum number of attempts')
        print('Please Retry')
    
            
```
